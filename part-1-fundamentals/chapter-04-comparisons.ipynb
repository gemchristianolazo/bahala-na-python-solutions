## 1. Write five comparisons for your own small store or personal budget. Try to make each one return either True or False.

```python
stock = 20
price = 15
cost = 5
budget = 500

print(budget < price * stock) # Is the budget less than the total price of our stocks?
print(budget > cost * stock)  # Is the budget greater than the total cost of our stocks?
print(price > cost) # Is the price higher than the cost?
print(price == cost) # Is the cost and the price the same?
print(price < cost) # Is the price less than the cost?

```
