## 1. Write three store helper functions:

- one that calculates a subtotal
- one that checks if an item is low stock
- one that prints a receipt line

```python
inventory = {
    "sardines": {"qty": 12, "price": 26},
    "candy": {"qty": 60, "price": 2},
    "instant noodles": {"qty": 15, "price": 17},
    "bread": {"qty": 0, "price": 8},
    "zesto": {"qty": 4, "price": 12},
    "softdrinks": {"qty": 3, "price": 15}
}
```

```python
def calculate_subtotal(qty, price):
  subtotal = qty * price
  return subtotal
```

```python
calculate_subtotal(3, 15)
```

```python
def is_low_stock(target, reorder_level=5):
  for item, detail in inventory.items():
    if item == target:
      if detail['qty'] == 0:
        print(f"{item} is out of stock")
      elif detail['qty'] < reorder_level:
        print(f"{item} is low stock")
      else:
        print(f"{item} is in stock")
```

```python
is_low_stock("zesto")
```

```python
def generate_receipt_line(item, qty, price):
  total = qty * price
  receipt_line = f"{item} x{qty} = {total}"
  return receipt_line
```

```python
print(generate_receipt_line("softdrinks", 2, 15))
```
