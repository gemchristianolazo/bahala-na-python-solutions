## Initialize inventory

```python
inventory = {
    "laundry soap": {"qty": 25, "price": 5},
    "cigarettes": {"qty": 60, "price": 2},
    "candy": {"qty": 32, "price": 1},
    "instant noodles": {"qty": 12, "price": 10},
    "softdrinks": {"qty": 24, "price": 7},
}
```

##1. Display inventory

```python
print("Inventory")
for item, details in inventory.items():
  print(f"""
  Item: {item}
  Quantity: {details["qty"]},
  Price: {details["price"]}
  """)
```

##2. Add new items

```python
item = input("Enter item to add: ")
qty = int(input(f"Enter quantity for {item}: "))
price = int(input(f"Enter price for {item}: "))
inventory[item] = {"qty": qty, "price": price}
print(f"{item} added to inventory" )
```

##3. Sell items (reduce quantity)

```python
item = input("Enter item to sell: ")
qty = int(input(f"Enter quantity to sell for {item}: "))
inventory[item]["qty"] -= qty
print(f"{qty} {item} sold!")
```


##4. Check if an item is in stock


```python
item = input("Enter item to check stock: ")
print(f"Stock for {item}: {inventory[item]["qty"]}")
```

##5. Show total inventory value (quantity × price)

```python
total_inventory_value = 0

for item, details in inventory.items():
  total_inventory_value += details["qty"] * details["price"]

print(f"Total inventory value: {total_inventory_value}")
```

# Side quests

##1. Add a low stock warning when quantity drops below 5.

```python
for item, details in inventory.items():
  if details["qty"] < 5:
    print(f"Stock alert: {item} is running out of stock (quantity: {details["qty"]})!")
```

##2. Add a feature to sort inventory by quantity (lowest first)

```python
sorted_inventory = sorted(inventory.items(), key=lambda item: item[1]["qty"], reverse=True)

print("Sorted Inventory")
for item, details in sorted_inventory:
  print(f"""
  Item: {item}
  Quantity: {details["qty"]},
  Price: {details["price"]}
  """)

print(sorted_inventory)
```

## 3. Create a "restock list" that shows items below a threshold


```python
items_below_threshold = []

for item, details in inventory.items():
  if details['qty'] < 5:
    items_below_threshold.append(item)

print("Restock list: ")
for item, details in inventory.items():
  if item in items_below_threshold:
    print(f"""
    Item: {item}
    Quantity: {details["qty"]},
    Price: {details["price"]}
    """)
```

## 4. Add prices to your inventory and calculate total store value

```python
item = input("Enter item to change price: ")
price = int(input(f"Enter new price for {item}: "))
inventory[item]["price"] = price
print(f"{item} price changed to {price}")

total_store_value = 0

for item, details in inventory.items():
  total_store_value += details["qty"] * details["price"]

print(f"Total store value: {total_store_value}")

```
