##1. Write three store rules using and, or, and not. Keep each rule to one line first, then try adding parentheses.



```python
stock = 4
quantity = 2
price = 50
subtotal = 350
is_member = True
is_open = True
is_holiday = False
```

```python
if is_open and subtotal > 300:
  print("10% discount on next purchase.")

if is_member or is_holiday:
  print("Discount applied!")
```

```python
is_open = False

if not is_open:
  print("Come back later.")
```

```python
# With parentheses
is_open = True

if (is_open and subtotal > 300) and is_member:
  print("3% cashback credited on your account.")

if (subtotal < quantity * price or stock < quantity) or (subtotal > quantity * price):
  print("Double check the order.")
```

```python
is_member = False

if (not is_holiday and not is_member) and subtotal > 0:
  print("Thank you for buying.")
```
