## 1. Turn the stock checker into a nightly restock report. Mark items as sold out, low stock, or okay.

```python
inventory = {
    "laundry soap": 3,
    "candy": 15,
    "rice": 12,
    "sardines": 0,
    "shampoo": 4,
    "stick o": 50
}

while True:
  print("1.) Check inventory")
  print("2.) Generate stock report")
  print("3.) Exit")

  choice = int(input("Enter choice: "))

  if choice == 1:
    print("Checking inventory...")
    for item, stock in inventory.items():
      print(f"""
      Item: {item}
      Stock: {stock}
      """)
  elif choice == 2:
    print("Generating stock report...")
    for item, stock in inventory.items():
      if stock == 0:
        print(f"""
        Item {item}: sold out
        """)
      elif stock < 5:
        print(f"""
        Item {item}: low stock
        """)
      else:
        print(f"""
        Item {item}: okay
        """)
  elif choice == 3:
    print("Exiting...")
    break
  else:
    print("Invalid choice.")
```
